```python
import neattext.functions as fn
```

```python
df['clean_title'] = df['course_title'].apply(fn.remove_stopwords)
```

```python
df['clean_title'] = df['clean_title'].apply(fn.remove_special_characters)
```

```python
df['clean_title'].iloc[1:11]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
cv_m=cv.fit_transform(df['clean_title'])
```

```python
cv_m.toarray()
```

```python
df_cv_w = pd.DataFrame(cv_m.toarray(),columns=cv.get_feature_names())
df_cv_w
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
cv_sim_m = cosine_similarity(cv_m)
cv_sim_m
```

```python
coures_index = pd.Series(df.index,index=df['course_title']).drop_duplicates()
```

```python
coures_index
```

```python
title = "How To Maximize Your Profits Trading Options"  
```

```python
index = coures_index[title]
```

```python
index
```

```python
cv_sim_m[index]
```

```python
scores = list(enumerate(cv_sim_m[index]))
scores
```

```python
stored_scores = sorted(scores,key=lambda x:x[1],reverse=True)
```

```python
stored_scores
```

```python
stored_indeces = [i[0] for i in stored_scores[1:]]
stored_indeces
```

```python
stored_values = [i[1] for i in stored_scores[1:]]
stored_values
```

```python
recommend_result = df.iloc[stored_indeces]
recommend_result
```
